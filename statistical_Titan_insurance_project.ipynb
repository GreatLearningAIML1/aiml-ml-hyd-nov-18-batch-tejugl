```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_1samp, ttest_ind
import math
from statsmodels.stats.power import ttest_power
```

```python
# IMPORTING DATA FROM THE EXCEL FILE.

excel_file = 'titan.xlsx'
dataset = pd.read_excel(excel_file,index_col=0)
dataset
```

```python
# TO FIND THE MEAN OF OLD_SCHEME.

dataset.loc[:,'OLD_SCHEME'].mean()
```

```python
# TO FIND THE MEAN OF NEW_SCHEME.

dataset.loc[:,'NEW_SCHEME'].mean()
```

# NULL HYPOTHESIS : There is no difference between the outputs of new & old schemes


# ALTERNATIVE HYPOTHESIS : There is a difference between the outputs of new & old schemes 

```python
Olddf = dataset['OLD_SCHEME(in thousands)']
Newdf = dataset['NEW_SCHEME(in thousands)']
```

```python
ttest = ttest_ind(Olddf, Newdf, axis=0, equal_var=False)
ttest
```

 p>0.05  Hence null hypothesis is accepted. There is no difference between the outputs of new and old scheme.

```python
from scipy.stats import ttest_rel
two_sample=ttest_rel(Olddf,Newdf)
two_sample
```

 Null Hypothesis : The average output has not increased by £5000 in the scheme compared to the old scheme, or mean difference is less than equal to 5000

Alternate Hypothesis : The average output has increased by £5000 in the scheme compared to the old scheme, or mean difference is greater than 5000


```python
dataset['diff']=dataset['NEW_SCHEME(in thousands)']-dataset['OLD_SCHEME(in thousands)']
from scipy.stats import ttest_1samp
one_sample = ttest_1samp(dataset['diff'], 5)
one_sample
```

 p>0.05. Null hypothesis is accepted.
The average output has not increased by £5000 in the new scheme compared to the old scheme

```python
TypeIError=1-0.14
print("The probability of making TypeIError is",TypeIError)
```

```python
effect=dataset['diff'].mean()/dataset['diff'].std()
Result =ttest_power(effect, nobs=30, alpha=0.86)
print("the power of the test is",Result)
```
