### Find out the no of male & female employees in the company

*Hint: Use can use a condition to find out the no of 1s and 0s in gender and sum it*

```python
print ('No . Of Male employees are : {}'.format(len(hr[(hr['Gender']==1)])))
```

```python
print ('No . Of Female employees are : {}'.format(len(hr[(hr['Gender']==2)])))
```

### Find out the IQR for the column - Age 

*Hint: IQR - Interquartile range = 75th percentile - 25th percentile* 

```python
hr['Age'].quantile(0.75)-hr['Age'].quantile(0.25)
```

### Find out the mean, median and variance for the column - DailyRate 

```python
print('Mean of Daily Rate :{}'.format(hr['DailyRate'].mean()))
print('Median of Daily Rate :{}'.format(hr['DailyRate'].median()))
print('Variance of Daily Rate :{}'.format(hr['DailyRate'].var()))
```

### Plot the attrition column (Since it is a categorical variable, use histogram to plot the data)

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(hr['Attrition'].value_counts())
plt.show()
```

```python
hr['Attrition'].value_counts().plot(kind='bar')
plt.xlabel('Attrition')
plt.ylabel('No . of Persons')
plt.show()
```

### Plot Total working years using matplotlib plot function

```python
plt.hist(hr['TotalWorkingYears'],bins=20,rwidth=0.7)
plt.xlabel('Work Experience in Years')
plt.ylabel('No . of Persons')
plt.show()
```
